```python
import dlib
import cv2
```

```python
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/Screenshot (2).png')
cv2_imshow(image)
```

# We use HOG and DLib for Face detection with OpenCV

```python
face_detector_dlib = dlib.get_frontal_face_detector()
```

```python
detections = face_detector_dlib(image)
```

```python
detections
```

```python
len(detections)
```

# Showing 4 detections were made 

```python
for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l,t), (r,b), (0,255, 255), 2)
  cv2_imshow(image)
  #print(face.left())
  #print(face.right())
  #print(face.top())
  #print(face.bottom())
```
